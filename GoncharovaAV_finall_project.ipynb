## Автоматизация расчета концентрации IgG к вирусу кори

### Гончарова А.В.

```python
# импорт библиотек и функций, режим отображения графиков 
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# создаем массивы y (концентрация) и x (ОП)
y = np.array([1, 1.93, 6.63, 10.23, 12.6, 23.7])
x = np.array([0.162, 0.277, 0.4105, 0.6545, 1.127, 2.0695])
n = 6
# расчет коэффициентов регрессии
b = (np.mean(x * y) - np.mean(x) * np.mean(y)) / (np.mean(x**2) - np.mean(x) ** 2)
a = np.mean(y) - b * np.mean(x)
y_pred = a + b * x
print(f'коэффициенты регрессии: a = {round(a, 2)}; b = {round(b, 2)}')
print()

# расчет коэффициента корреляции r
r = b * np.std(x) / np.std(y)
print(f'коэффициент корреляции: r = {round(r, 4)}')
print(f'Выявили прямую сильную корреляционную связь между ОП и концентрацией IgG к вирусу кори')
print()
print('Модель линейной регрессии представлена уравнением:')
print(f'концентрация = 0,38 + 11,45 * ОП')
print()

# расчет коэффициента детерминации R2
R2 = r**2
print(f'коэффициент детерминации: R2 = {round(R2, 4)}')
print(f'96,36% вариации концентрации IgG (y) объясняется вариацией фактора x – оптическая плотность.')
print()

# расчет F-критерия Фишера
F_fact = (r**2 * (n - 2)) / (1 - r**2)
F_fact
print(f'F-критерий Фишера: F_fact = {round(F_fact, 2)}')
print('Так как  Fфакт = 105,92 > Fкр = 7,71, то уравнение регрессии статистически значимо.')
print()

# расчет t-статистик Стьдента для коэффициентов регрессии
s_residual = np.sqrt(np.sum((y - y_pred)**2) / (n - 2))
m_a = s_residual * np.sqrt(np.sum(x ** 2)) / (n * np.std(x))
m_b = s_residual / (np.std(x) * np.sqrt(n))
t_a = a / m_a
t_b = b / m_b
print(f't-статистика Стьюдента: t_a = {round(t_a, 2)}, t_b = {round(t_b, 2)}')
print(f'коэффициент а статистически не значим, коэффициент b статистически значим')
print()
print('Построенная модель линейной регрессии для расчета концентрации IgG к вирусу кори признана статистически значимой')
print('и может использоваться в практической работе.')
print()
# расчет концентрации для образцов пациентов
m = int(input('Введите число исследуемых образцов: '))
N = []
for i in range(1, m + 1):
    fio = input(f'"Введите ID образца" {i}:')
    N.append(fio)
for j in N:
    a1 = float(input(f'Введите ОП образца {j}: '))
    a2 = float(input(f'Введите ОП образца {j}: '))
    OP = (a1 + a2) / 2
    Conc = a + b * OP
    print(f'Концентрация IgG (образец {j}) = {round(Conc, 2)} ME/мл')
    print()
```
